```python
# Количество записей каждого события
gh.groupby(by="event_type")["event_type"].count()
```

```python
# Ненулевые столбцы по каждому событию

events = gh["event_type"].unique()
not_null_columns_by_events = {}

for event in events:
    not_null_columns_by_events[event] = gh.columns[gh[gh["event_type"] == event].any()]

not_null_columns_by_events
```
