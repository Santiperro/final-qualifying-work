```python
from service_connector import ServiceConnector
import asyncio
import nest_asyncio
nest_asyncio.apply()


service_connector = ServiceConnector()
loop = asyncio.get_event_loop()

# Запрос данных о 4000 репозиторев за 4 месяца, которые имеют минимум 100 звезд
transactions_dict = ['pushes'
                    'avg_push_size',
                    'pull_requests',
                    'merged_pull_requests_ratio',
                    'issues',
                    'closed_issues_ratio',
                    'watches',
                    'forks',
                    'new_members',
                    'language',
                    'license_name',
                    'is_deleted_or_private']

github_data = loop.run_until_complete(service_connector.get_data_from_services(
                                                                 transactions_dict,
                                                                 "2023-01-01", 
                                                                 "2023-06-01", 
                                                                 limit=1000,
                                                                 min_watch_event_count=100,
                                                                 min_new_members_count=2,
                                                                 is_new_repos=False))
github_data
```

```python
from github_data_converter import GithubDataConverter
from pattern_miner import PatternMiner


# Конвертирование в транзакции
github_data_converter = GithubDataConverter()
transactions = github_data_converter.convert_data_to_transactions(github_data,                                                         
        {'pushes': 'dec',
        'avg_push_size': 'qua',
        'pull_requests': 'dec',
        'merged_pull_requests_ratio': 'dec',
        'issues': 'dec',
        'closed_issues_ratio': 'dec',
        'watches': 'dec',
        'forks': 'dec',
        'new_members': 'dec'})
transactions
```

```python
# Поиск шаблонов
pattern_miner = PatternMiner()
patterns = pattern_miner.mine_patterns(transactions, 0.01, 0.1, 1)
patterns
```

```python
# Сохранение в excel
patterns.to_excel("patterns.xlsx")
```
